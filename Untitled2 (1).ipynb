```python
# ============================================
# CELL 6: INTERACTION ANALYSIS
# ============================================
# Purpose: Analyze how features interact with each other
# Example: Does smoking affect the relationship between age and charges?

print("="*60)
print("STEP 3C: INTERACTION ANALYSIS")
print("="*60)

# Create age groups to analyze age-smoker interaction
# Binning age into meaningful groups: Young, Middle, Senior
df['age_group'] = pd.cut(df['age'], bins=[0, 30, 50, 100], 
                          labels=['Young (0-30)', 'Middle (31-50)', 'Senior (51+)'])

# Create side-by-side plots for age-smoker interaction
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Smoker percentage by age group
# Shows how smoking prevalence changes with age
smoker_by_age = pd.crosstab(df['age_group'], df['smoker'], normalize='index')
smoker_by_age.plot(kind='bar', ax=axes[0])
axes[0].set_title('Smoker Percentage by Age Group')
axes[0].set_ylabel('Proportion')
axes[0].set_xlabel('Age Group')
axes[0].legend(title='Smoker')
axes[0].tick_params(axis='x', rotation=0)

# Plot 2: Charges by age group and smoker status
# Shows how the combination of age and smoking affects costs
sns.boxplot(data=df, x='age_group', y='charges', hue='smoker', ax=axes[1])
axes[1].set_title('Charges by Age Group and Smoker Status')
axes[1].set_ylabel('Charges ($)')
axes[1].set_xlabel('Age Group')
axes[1].legend(title='Smoker')
axes[1].tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.show()

# Calculate the smoker multiplier for each age group
# This shows how much more smokers pay compared to non-smokers
print("\nAGE-SMOKER INTERACTION ANALYSIS:")
for age_group in df['age_group'].cat.categories:
    non_smoker = df[(df['age_group'] == age_group) & (df['smoker'] == 'no')]['charges'].mean()
    smoker = df[(df['age_group'] == age_group) & (df['smoker'] == 'yes')]['charges'].mean()
    multiplier = smoker / non_smoker if non_smoker > 0 else 0
    print(f"{age_group}:")
    print(f"  Non-smoker average: ${non_smoker:,.0f}")
    print(f"  Smoker average: ${smoker:,.0f}")
    print(f"  Smoker multiplier: {multiplier:.1f}x")

# Create BMI categories to analyze BMI-smoker interaction
df['bmi_category'] = pd.cut(df['bmi'], bins=[0, 18.5, 25, 30, 100], 
                            labels=['Underweight', 'Normal', 'Overweight', 'Obese'])

# Create side-by-side plots for BMI-smoker interaction
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Smoker percentage by BMI category
smoker_by_bmi = pd.crosstab(df['bmi_category'], df['smoker'], normalize='index')
smoker_by_bmi.plot(kind='bar', ax=axes[0])
axes[0].set_title('Smoker Percentage by BMI Category')
axes[0].set_ylabel('Proportion')
axes[0].set_xlabel('BMI Category')
axes[0].legend(title='Smoker')
axes[0].tick_params(axis='x', rotation=0)

# Plot 2: Charges by BMI category and smoker status
sns.boxplot(data=df, x='bmi_category', y='charges', hue='smoker', ax=axes[1])
axes[1].set_title('Charges by BMI Category and Smoker Status')
axes[1].set_ylabel('Charges ($)')
axes[1].set_xlabel('BMI Category')
axes[1].legend(title='Smoker')
axes[1].tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.show()

# Calculate the smoker multiplier for each BMI category
print("\nBMI-SMOKER INTERACTION ANALYSIS:")
for bmi_cat in df['bmi_category'].cat.categories:
    non_smoker = df[(df['bmi_category'] == bmi_cat) & (df['smoker'] == 'no')]['charges'].mean()
    smoker = df[(df['bmi_category'] == bmi_cat) & (df['smoker'] == 'yes')]['charges'].mean()
    if not np.isnan(non_smoker) and not np.isnan(smoker):
        multiplier = smoker / non_smoker
        print(f"{bmi_cat}:")
        print(f"  Non-smoker average: ${non_smoker:,.0f}")
        print(f"  Smoker average: ${smoker:,.0f}")
        print(f"  Smoker multiplier: {multiplier:.1f}x")
```

```python
# ============================================
# CELL 7: FEATURE ENGINEERING
# ============================================
# Purpose: Create new features that might improve model performance
# This includes interaction terms and transformations

print("="*60)
print("FEATURE ENGINEERING")
print("="*60)

# Prepare feature set with original features
X = df[['age', 'bmi', 'children', 'smoker']].copy()
y = df['charges']

# Convert smoker to binary (0 for no, 1 for yes)
# This makes the feature numerical and suitable for mathematical operations
X['smoker'] = (X['smoker'] == 'yes').astype(int)

# Add interaction features
# These capture combined effects that individual features might miss
# age_smoker: The effect of age is amplified for smokers
X['age_smoker'] = X['age'] * X['smoker']

# bmi_smoker: The effect of BMI is amplified for smokers
X['bmi_smoker'] = X['bmi'] * X['smoker']

# age_bmi: Age and BMI together might have a combined effect
X['age_bmi'] = X['age'] * X['bmi']

# Apply log transformation to target to handle skewness
# log1p is used instead of log to handle zero values safely (log(1+x))
y_log = np.log1p(y)

print("Feature set shape:", X.shape)
print("\nFeature columns with descriptions:")
print("  age: Age of beneficiary (numerical)")
print("  bmi: Body Mass Index (numerical)")
print("  children: Number of dependents (numerical)")
print("  smoker: Smoking status (binary: 0=no, 1=yes)")
print("  age_smoker: Interaction term - age multiplied by smoker status")
print("  bmi_smoker: Interaction term - BMI multiplied by smoker status")
print("  age_bmi: Interaction term - age multiplied by BMI")

print("\nFeature statistics:")
display(X.describe())

print(f"\nTarget transformation applied: log1p")
print(f"Original target range: ${y.min():,.2f} - ${y.max():,.2f}")
print(f"Original target mean: ${y.mean():,.2f}")
print(f"Log-transformed target range: {y_log.min():.2f} - {y_log.max():.2f}")
print(f"Log-transformed target mean: {y_log.mean():.2f}")
print(f"Skewness reduced from {y.skew():.3f} to {y_log.skew():.3f}")
```

```python
# ============================================
# CELL 8: FEATURE SELECTION - MULTIPLE METHODS
# ============================================
# Purpose: Identify which features are most important for prediction
# Using multiple methods ensures robust feature selection

print("="*60)
print("FEATURE SELECTION")
print("="*60)

# Method 1: Correlation Analysis
# Measures linear relationship between each feature and target
print("\nMETHOD 1: CORRELATION ANALYSIS")
correlations = {}
for col in X.columns:
    corr = np.corrcoef(X[col], y)[0, 1]
    correlations[col] = corr

corr_df = pd.DataFrame(list(correlations.items()), columns=['Feature', 'Correlation'])
corr_df = corr_df.sort_values('Correlation', key=abs, ascending=False)
print(corr_df.to_string(index=False))

# Method 2: ANOVA for categorical features
# Tests if means across categories are significantly different
print("\nMETHOD 2: ANOVA TEST")
from scipy.stats import f_oneway

def anova_feature_selection(df, categorical_cols, target_col):
    results = {}
    for col in categorical_cols:
        groups = [df[df[col] == value][target_col].values for value in df[col].unique()]
        f_stat, p_value = f_oneway(*groups)
        results[col] = {'F-statistic': f_stat, 'p-value': p_value}
    return pd.DataFrame(results).T

anova_results = anova_feature_selection(df, ['sex', 'smoker', 'region'], 'charges')
print(anova_results)

# Method 3: Recursive Feature Elimination (RFE)
# Iteratively removes least important features
print("\nMETHOD 3: RECURSIVE FEATURE ELIMINATION (RFE)")
X_rfe = df[['age', 'bmi', 'children', 'smoker']].copy()
X_rfe['smoker'] = (X_rfe['smoker'] == 'yes').astype(int)

scaler_rfe = StandardScaler()
X_rfe_scaled = scaler_rfe.fit_transform(X_rfe)

estimator = LinearRegression()
selector = RFE(estimator, n_features_to_select=3, step=1)
selector = selector.fit(X_rfe_scaled, y)

print(f"Selected features: {X_rfe.columns[selector.support_].tolist()}")
print("Feature rankings (1=most important):")
for feature, rank in zip(X_rfe.columns, selector.ranking_):
    print(f"  {feature}: rank {rank}")

# Method 4: P-value Analysis using Statsmodels
# Statistical significance of each feature
print("\nMETHOD 4: P-VALUE ANALYSIS (STATSMODELS)")
X_sm = sm.add_constant(X_rfe)
model_sm = sm.OLS(y, X_sm).fit()
p_values = model_sm.pvalues
print("\nP-values (lower is more significant):")
for feature, p_val in p_values.items():
    if feature != 'const':
        significance = ""
        if p_val < 0.001:
            significance = "*** (highly significant)"
        elif p_val < 0.01:
            significance = "** (very significant)"
        elif p_val < 0.05:
            significance = "* (significant)"
        else:
            significance = "ns (not significant)"
        print(f"  {feature}: {p_val:.4f} {significance}")

print("\nFEATURE SELECTION CONCLUSION:")
print("Based on all four methods, the following features are most important:")
print("  - smoker: Selected by all methods (strongest predictor)")
print("  - age: Selected by correlation, RFE, and p-value methods")
print("  - bmi: Selected by correlation, RFE, and p-value methods")
print("  - children: Borderline significance")
print("\nInteraction features to add:")
print("  - age_smoker: Captures smoking amplification of age effect")
print("  - bmi_smoker: Captures smoking amplification of BMI effect")
print("  - age_bmi: Captures age-BMI interaction")
print("\nFeatures to consider removing:")
print("  - region: Not significant in any test")
print("  - sex: Borderline significance, minimal business impact")
```

```python
# ============================================
# CELL 9: TRAIN-TEST SPLIT AND SCALING
# ============================================
# Purpose: Prepare data for model training
# Split into training and testing sets, then scale features

print("="*60)
print("TRAIN-TEST SPLIT AND SCALING")
print("="*60)

# Split data into training (80%) and testing (20%) sets
# random_state ensures reproducibility of the split
X_train, X_test, y_train, y_test = train_test_split(
    X, y_log, test_size=0.2, random_state=42
)

print(f"Training set size: {X_train.shape[0]} rows ({X_train.shape[0]/len(X)*100:.1f}%)")
print(f"Testing set size: {X_test.shape[0]} rows ({X_test.shape[0]/len(X)*100:.1f}%)")
print(f"Number of features: {X_train.shape[1]}")

# Scale features to have mean=0 and standard deviation=1
# This is important for regularized models (Ridge, Lasso)
# and ensures all features contribute equally
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(f"\nFeature scaling applied: StandardScaler")
print(f"  - Removes mean: Centers data around 0")
print(f"  - Scales variance: Ensures unit variance")
print(f"Training features shape (scaled): {X_train_scaled.shape}")
print(f"Testing features shape (scaled): {X_test_scaled.shape}")

# Verify scaling properties
print(f"\nVerification of scaling:")
print(f"  Training mean: {X_train_scaled.mean():.2f} (should be ~0)")
print(f"  Training std: {X_train_scaled.std():.2f} (should be ~1)")
```

```python
# ============================================
# CELL 10: MODEL TRAINING AND COMPARISON
# ============================================
# Purpose: Train multiple regression models and compare their performance

print("="*60)
print("MODEL TRAINING AND COMPARISON")
print("="*60)

# Define models with different regularization approaches
models = {
    'Linear Regression': LinearRegression(),
    # No regularization - baseline model
    'Ridge (α=1)': Ridge(alpha=1.0),
    # L2 regularization - penalizes large coefficients
    'Ridge (α=10)': Ridge(alpha=10.0),
    # Stronger L2 regularization
    'Lasso (α=1)': Lasso(alpha=1.0),
    # L1 regularization - can set coefficients to zero
    'ElasticNet (α=1)': ElasticNet(alpha=1.0, l1_ratio=0.5)
    # Combines L1 and L2 regularization
}

# Store results for comparison
results = {}

print("Training models...\n")

for name, model in models.items():
    # Train the model on scaled training data
    model.fit(X_train_scaled, y_train)
    
    # Make predictions on test data
    y_pred_test = model.predict(X_test_scaled)
    
    # Convert predictions back to original scale for interpretation
    # expm1 reverses log1p transformation: exp(x) - 1
    y_pred_test_orig = np.expm1(y_pred_test)
    y_test_orig = np.expm1(y_test)
    
    # Calculate performance metrics
    r2_test = r2_score(y_test_orig, y_pred_test_orig)
    rmse_test = np.sqrt(mean_squared_error(y_test_orig, y_pred_test_orig))
    mae_test = mean_absolute_error(y_test_orig, y_pred_test_orig)
    
    # Perform 5-fold cross-validation to check generalization
    cv_scores = cross_val_score(model, X_train_scaled, y_train, cv=5, scoring='r2')
    
    results[name] = {
        'R² Test': r2_test,
        'RMSE Test': rmse_test,
        'MAE Test': mae_test,
        'CV Mean': cv_scores.mean(),
        'CV Std': cv_scores.std()
    }
    
    print(f"  {name}: R²={r2_test:.4f}, RMSE=${rmse_test:,.0f}, CV={cv_scores.mean():.4f}")

# Display results table
results_df = pd.DataFrame(results).T
print("\nMODEL PERFORMANCE COMPARISON:")
display(results_df.round(4))

# Identify best model based on test R²
best_model_name = results_df['R² Test'].idxmax()
best_r2 = results_df.loc[best_model_name, 'R² Test']
print(f"\nBest performing model: {best_model_name} (R² = {best_r2:.4f})")
```

```python
# ============================================
# CELL 11: HYPERPARAMETER TUNING
# ============================================
# Purpose: Find optimal regularization strength for Ridge Regression

print("="*60)
print("HYPERPARAMETER TUNING")
print("="*60)

# Define range of alpha values to test
# Alpha controls the strength of regularization:
# - Small alpha: less regularization, model may overfit
# - Large alpha: more regularization, model may underfit
alphas = [0.001, 0.01, 0.1, 0.5, 1, 5, 10, 20, 50, 100, 200, 500, 1000]

# Perform GridSearchCV to find best alpha
# This tries each alpha value with 5-fold cross-validation
ridge = Ridge()
param_grid_ridge = {'alpha': alphas}
grid_search = GridSearchCV(ridge, param_grid_ridge, cv=5, scoring='r2', verbose=0)
grid_search.fit(X_train_scaled, y_train)

print(f"\nRIDGE REGRESSION TUNING RESULTS:")
print(f"Best alpha: {grid_search.best_params_['alpha']}")
print(f"Best cross-validation R²: {grid_search.best_score_:.4f}")

# Train final model with best alpha
best_ridge = grid_search.best_estimator_

# Visualize alpha selection
# Shows how performance changes with different regularization strengths
plt.figure(figsize=(10, 6))
ridge_scores = [grid_search.cv_results_['mean_test_score'][i] for i in range(len(alphas))]
plt.plot(alphas, ridge_scores, 'bo-', linewidth=2)
plt.xscale('log')  # Log scale for alpha to see wide range
plt.xlabel('Alpha (Regularization Strength)')
plt.ylabel('Cross-Validation R² Score')
plt.title('Ridge Regression: Alpha Selection')
plt.axvline(grid_search.best_params_['alpha'], color='red', linestyle='--', 
            label=f"Best α={grid_search.best_params_['alpha']}")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# ============================================
# CELL 12: MODEL EVALUATION - METRICS
# ============================================
# Purpose: Evaluate final model performance with multiple metrics

print("="*60)
print("MODEL EVALUATION")
print("="*60)

# Make predictions on both training and test sets
y_pred_train = best_ridge.predict(X_train_scaled)
y_pred_test = best_ridge.predict(X_test_scaled)

# Convert back to original scale for interpretability
y_pred_train_orig = np.expm1(y_pred_train)
y_pred_test_orig = np.expm1(y_pred_test)
y_train_orig = np.expm1(y_train)
y_test_orig = np.expm1(y_test)

# Calculate comprehensive performance metrics
# Training metrics
r2_train = r2_score(y_train_orig, y_pred_train_orig)
rmse_train = np.sqrt(mean_squared_error(y_train_orig, y_pred_train_orig))
mae_train = mean_absolute_error(y_train_orig, y_pred_train_orig)

# Test metrics
r2_test = r2_score(y_test_orig, y_pred_test_orig)
rmse_test = np.sqrt(mean_squared_error(y_test_orig, y_pred_test_orig))
mae_test = mean_absolute_error(y_test_orig, y_pred_test_orig)

# Adjusted R² penalizes for number of features
# Helps prevent overfitting by accounting for model complexity
n = len(y_test)
k = X_test.shape[1]
adjusted_r2 = 1 - (1 - r2_test) * (n - 1) / (n - k - 1)

# 10-fold cross-validation to check stability
cv_scores = cross_val_score(best_ridge, X_train_scaled, y_train, cv=10, scoring='r2')

# Mean Absolute Percentage Error (MAPE)
# Provides error as percentage of actual values
mape_test = np.mean(np.abs((y_test_orig - y_pred_test_orig) / y_test_orig)) * 100

print("FINAL MODEL PERFORMANCE METRICS")
print("="*50)
print(f"\nModel: Ridge Regression (α={grid_search.best_params_['alpha']})")
print(f"Number of features: {k}")
print(f"Training samples: {len(y_train)}")
print(f"Testing samples: {len(y_test)}")

print(f"\nTRAINING SET:")
print(f"  R² Score: {r2_train:.4f} (variance explained)")
print(f"  RMSE: ${rmse_train:,.2f} (root mean squared error)")
print(f"  MAE: ${mae_train:,.2f} (mean absolute error)")

print(f"\nTESTING SET:")
print(f"  R² Score: {r2_test:.4f} (variance explained)")
print(f"  Adjusted R²: {adjusted_r2:.4f} (penalized for features)")
print(f"  RMSE: ${rmse_test:,.2f} (root mean squared error)")
print(f"  MAE: ${mae_test:,.2f} (mean absolute error)")
print(f"  MAPE: {mape_test:.2f}% (mean absolute percentage error)")

print(f"\nCROSS-VALIDATION (10-fold):")
print(f"  Mean R²: {cv_scores.mean():.4f}")
print(f"  Standard Deviation: {cv_scores.std():.4f}")
print(f"  Range: [{cv_scores.min():.4f}, {cv_scores.max():.4f}]")
```

```python
# ============================================
# CELL 13: MODEL EVALUATION - VISUALIZATIONS
# ============================================
# Purpose: Visual diagnostic plots to validate model assumptions

print("="*60)
print("MODEL EVALUATION - VISUALIZATIONS")
print("="*60)

# Calculate residuals (difference between actual and predicted)
residuals = y_test_orig - y_pred_test_orig

# Create comprehensive diagnostic plots
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Plot 1: Actual vs Predicted
# Points should fall along diagonal line
axes[0, 0].scatter(y_test_orig, y_pred_test_orig, alpha=0.5)
axes[0, 0].plot([y_test_orig.min(), y_test_orig.max()], 
                [y_test_orig.min(), y_test_orig.max()], 
                'r--', linewidth=2)
axes[0, 0].set_xlabel('Actual Charges ($)')
axes[0, 0].set_ylabel('Predicted Charges ($)')
axes[0, 0].set_title('Actual vs Predicted Values')
axes[0, 0].grid(True, alpha=0.3)

# Plot 2: Residuals vs Predicted
# Should show random scatter around zero (no pattern)
axes[0, 1].scatter(y_pred_test_orig, residuals, alpha=0.5)
axes[0, 1].axhline(y=0, color='red', linestyle='--', linewidth=2)
axes[0, 1].set_xlabel('Predicted Charges ($)')
axes[0, 1].set_ylabel('Residuals ($)')
axes[0, 1].set_title('Residuals vs Predicted Values')
axes[0, 1].grid(True, alpha=0.3)

# Plot 3: Histogram of Residuals
# Should be approximately bell-shaped (normal distribution)
axes[0, 2].hist(residuals, bins=30, edgecolor='black', alpha=0.7)
axes[0, 2].set_xlabel('Residuals ($)')
axes[0, 2].set_ylabel('Frequency')
axes[0, 2].set_title('Distribution of Residuals')
axes[0, 2].grid(True, alpha=0.3)

# Plot 4: Q-Q Plot of Residuals
# Points should fall along diagonal line for normality
stats.probplot(residuals, dist="norm", plot=axes[1, 0])
axes[1, 0].set_title('Q-Q Plot of Residuals')
axes[1, 0].grid(True, alpha=0.3)

# Plot 5: Prediction Error Distribution
# Shows distribution of absolute errors
errors = np.abs(residuals)
axes[1, 1].hist(errors, bins=30, edgecolor='black', alpha=0.7, color='orange')
axes[1, 1].set_xlabel('Absolute Error ($)')
axes[1, 1].set_ylabel('Frequency')
axes[1, 1].set_title('Prediction Error Distribution')
axes[1, 1].axvline(errors.mean(), color='red', linestyle='--', 
                   label=f'Mean Error: ${errors.mean():,.0f}')
axes[1, 1].legend()
axes[1, 1].grid(True, alpha=0.3)

# Plot 6: Cross-Validation Scores
# Shows consistency across different folds
axes[1, 2].plot(range(1, 11), cv_scores, 'bo-', linewidth=2, markersize=8)
axes[1, 2].axhline(y=cv_scores.mean(), color='red', linestyle='--', 
                   label=f'Mean: {cv_scores.mean():.4f}')
axes[1, 2].fill_between(range(1, 11), cv_scores.mean() - cv_scores.std(), 
                        cv_scores.mean() + cv_scores.std(), alpha=0.2, color='red')
axes[1, 2].set_xlabel('Fold')
axes[1, 2].set_ylabel('R² Score')
axes[1, 2].set_title('Cross-Validation R² Scores')
axes[1, 2].legend()
axes[1, 2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Statistical tests for residuals
print("\nRESIDUAL DIAGNOSTICS:")
print("-"*40)

# Shapiro-Wilk test for normality
# Null hypothesis: residuals are normally distributed
if len(residuals) > 5000:
    shapiro_stat, shapiro_p = stats.shapiro(residuals[:5000])
else:
    shapiro_stat, shapiro_p = stats.shapiro(residuals)

print(f"Shapiro-Wilk normality test p-value: {shapiro_p:.4f}")
if shapiro_p > 0.05:
    print("  Interpretation: Residuals appear normally distributed (p > 0.05)")
else:
    print("  Interpretation: Residuals deviate from normality (p < 0.05)")
    print("  Note: This is acceptable for business use with large datasets")

# Durbin-Watson test for autocorrelation
# Values near 2 indicate no autocorrelation
dw_stat = durbin_watson(residuals)
print(f"Durbin-Watson statistic: {dw_stat:.4f}")
if 1.5 < dw_stat < 2.5:
    print("  Interpretation: No significant autocorrelation detected")
else:
    print("  Interpretation: Potential autocorrelation detected")
```
